## Notation

Physicists use a variety of different notations so lets start by reviewing some notation from intro physics and possibly introducing some new notation. We should all be familiar with vector notation at this point but lets just go over it anyway. 

### Vector Notation
A vector in physics is some quantity that has both a magnitude (size) and direction. The math and computer science people have their own definitions of vectors which may occasionally be referenced in these notes but usually we will use the physics definition. If either of the other definitions are used they will be addressed and defined at that time. 

A vector is usually denoted either as a lower case bold faced letter $\bf{x}$ or as a lower case letter with an arrow above it $\vec{x}$. Or sometimes as both $\bf{\vec{x}}$. In my notes I will usually use the arrow notation. 

The most basic vector is a position vector. It is simply the coordinates of a particle in 3D space and thus has 3 parts. $\vec{x} = (x, y, z)$ or sometimes we will use unit vector notation which uses a hat $\hat{}$ to denote *unit vectors* which are vectors of magnitude 1. Unit vectors are really just pointers that point in a certain direction. $\vec{x} = x\hat{i} + y\hat{j} + z\hat{k}$. Sometimes we wont use $x, y, z$ but instead we will denote the coordinates with subscripts such as $\vec{x} = (x_1, x_2, x_3) = x_1\hat{i} + x_2\hat{j} + x_3\hat{k}$ and sometimes we will use subscripts on the letter $e$ to denote unit vectors $\vec{x} = x_1\hat{e_1} + x_2\hat{e_2} + x_3\hat{e_3}$. 

This final notation will become extremely important and useful when we start working more with tensors and transformations later on and in other advanced physics classes.

### Differentiating Vectors

If $\vec{x}$ is the position of a particle then the derivative of $\vec{x}$ with respect to time is denoted by $\frac{d\vec{x}}{dt}$ and is the particles velocity which we will assign to the letter $v$ or in vector notation $\vec{v} = \frac{d\vec{x}}{dt} = (\frac{dx}{dt}, \frac{dy}{dt}, \frac{dz}{dt})$. In classical mechanics we will often use a shorthand notation for derivatives with respect to time $\dot{\vec{x}}$. 

If you again differentiate $\vec{x}$ with respect to time, so $\frac{d\vec{v}}{dt} = \frac{d^2\vec{x}}{dt^2}$ that is a particles acceleration $\vec{a}$. Using dot notation for the second time derivative then is $\vec{a} = \dot{\vec{v}} = \ddot{\vec{x}}$

All the above notation assumes cartesian coordinates which we will often use. Often though it is more useful to use cylindrical or spherical coordinates depending on the symmetry of the system we are dealing with. In those cases (and often even when using cartesain coordinates) our position vector will be denoted by $\vec{r}$.

In cylindrical coordinates <!-- TODO: (check this) --> $\vec{r} = (\rho, \phi, z)$ and in spherical coordinates $\vec{r} = (r, \phi, \theta)$ 

Since we're in jupyter notebooks lets plot the difference between cylindrical (or polar in 2D) and cartesian coordinates. 

```python
# First we need to import the necessary packages
import numpy as np
from matplotlib import pyplot as plt
```

```python
fig, ax = plt.subplots(1, 2, figsize=(9, 6))
fig.suptitle("Position vector $\\vec{r}$", fontsize=20)
ax[0].arrow(0, 0, 3, 3, head_width=0.2)
ax[0].set_xlim(-5, 5)
ax[0].set_ylim(-5, 5)
ax[0].hlines(0, -5, 5, linestyles="--", colors="k")
ax[0].hlines(3, 0, 3, linestyles=":", colors="r")
ax[0].vlines(0, -5, 5, linestyles="--", colors="k")
ax[0].vlines(3, 3, 0, linestyles=":", colors="r")
ax[0].annotate("(3, 3)", (3.25, 3.25))
ax[0].set_title("Cartesian Coordinates")
ax[0].set_aspect("equal")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

theta = np.linspace(0, np.pi / 4, 20)
x = 2 * np.cos(theta)
y = 2 * np.sin(theta)

ax[1].arrow(0, 0, 3, 3, head_width=0.2)
ax[1].plot(x, y, ls=":", c="r")
ax[1].set_xlim(-5, 5)
ax[1].set_ylim(-5, 5)
ax[1].hlines(0, -5, 5, linestyles="--", colors="k")
ax[1].vlines(0, -5, 5, linestyles="--", colors="k")
ax[1].set_title("Polar Coordinates")
ax[1].set_aspect("equal")
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")
ax[1].annotate("$\\theta = \\frac{\pi}{4} = 45^\\circ$", (2.25, 1))
r = np.round(np.sqrt(3**2 + 3**2), 2)
ax[1].text(x=0.5, y=1, s=f"$r = {r}$", rotation=45)
fig.tight_layout()
plt.show()
```

Ok so that's a lot of code but, most of it is just formatting stuff. To actually plot the vector itself is very simple just use `plt.arrow(x0, y0, dx, dy)`
